```python
import json

```

```python
def load_data(filename):

    with open (filename,"r") as file :
        data = json.load(file)
        return data
def displaydata(data):
    print("Users Information :-")
    for user in data["users"]:
         print(f" user's name = {user['name']} user's ID = {user['id']} user's friends = {user['friends']} liked pages by user = {user['liked_pages']}")
    print("Liked page by user's information")
    for page in data["pages"]:
        print(f"user's id :- {page['id']} their laked pages about :- {page['name']} ")



```

```python
data = load_data("data.json")
```
