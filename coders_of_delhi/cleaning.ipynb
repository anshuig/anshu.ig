```python
import json
def opendata2(filename):
    with open (filename , "r") as file:
    
       data = json.load(file)
    return data
def clean_data(data):
# to remove non name user:-    
    data["users"] = [user for user in data["users"] if user["name"].strip()]
# to remove duplicates pehle humne firstly we convertes list into set uske badh humne set ko list mai convert kra     
    for user in data["users"]:
        user["friends"] = list(set(user["friends"]))

# list [____] kai andar agar "if" ka use ho rha hain toh iska matlab hain ki if badh wali conditions honi chahiye aagar nhi use toh uspe 
# for loop mat chalaoo
    data["users"] =  [ user for user in data["users"] if user["friends"] or user["liked_pages"]]

# dictionary mai ek key ki ek hi value hoti hain aagar koi key duplicate ho to
    unique_pages = {}
    for page in data["pages"]:
        unique_pages[page['id']] = page
        data["pages"] = list(unique_pages.values())

# humne unique pages naam ki dictionary banayi hain jisme duplicate page repeat nhi honge . lakin data['pages'] mai haame unique page dalna hoga 
# lakin json.file mai bas list hoti naaki dictionary toh hun list( ) ka use krke dictionary ko list mai convert krege . aagar hum aase hi 
# print karwayenge toh keys hi jayegi toh haame ".values()"  ka use krna hoga


data = opendata2("data2.json")
clean_data(data)
json.dump(data  ,open("cleaned_data2.json" , "w"), indent = 4 )
```
