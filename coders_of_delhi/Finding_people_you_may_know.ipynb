```python
import json

def laoo_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
        return data
def find_people_you_may_now(user_id,data):
    user_friends = {}
    for user in data['users']:
        user_friends[user["id"]] = set(user['friends'])
    if user_id not in user_friends:
        return []
    direct_friend = user_friends[user_id]
    suggestion = {}
    for friend in direct_friend:
        for mutual in user_friends[friend]:
# yaha pr humne mutual mai - friends[joki direct friends sai aaye thhe "2,3"] kai friends liye hain and usme sai phir mutuals nikalege
            if mutual != user_id and mutual  not in direct_friend:
                suggestion[mutual] = suggestion.get(mutual,0)+1
    sorted_suggestion = sorted(suggestion.items() , lambda x: x[1], reverse = True )
    return [ user_id for user_id ,  _ in sorted_suggestion]
                
        

data = laoo_data("data.json")
user_id = 1
recommandation = find_people_you_may_now(user_id,data)
print(recommandation)
        
```
